```python
import cv2
import numpy as np
import onnxruntime

classes = []

# 클래스 라벨 불러오기
with open("D:/InsectData/convertData/HighDefinition_data/classes.txt", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# ONNX 런타임 세션 생성
sess = onnxruntime.InferenceSession("D:/InsectData/data/model.onnx")

# 이미지 불러오기
image = cv2.imread("D:/InsectData/data/Validation/rawData/VS1/PtecticusTenebrifer/Caterpillar/HighDefinition/Capture_20211117_13_45_52.jpg")
image = cv2.resize(image, None, fx=0.4, fy=0.4)
height, width, channels = image.shape
colors = np.random.uniform(0, 255, 3)

# 이미지 전처리
blob = cv2.dnn.blobFromImage(image, 0.00392, (416, 416), (0, 0, 0), swapRB=True, crop=False)
blob = blob.transpose((0, 1, 2, 3))  # ONNX는 CHW가 아니라 HWC 포맷을 기대

# 모델 실행 및 출력 얻기
input_name = sess.get_inputs()[0].name
output_name = sess.get_outputs()[0].name
outs = sess.run([output_name],{input_name: blob})

print(input_name)
print(output_name)
print(np.array(outs).shape)

for i, out in enumerate(outs):
    print(f"Output node {i} shape: {out.shape}")
    print(out)

# 결과 처리
class_ids = []
confidences = []
boxes = []

# 각 출력 노드에 대해 반복
for out in outs:
    # 첫 번째 차원(1)은 배치 크기를 나타내므로 우리는 이를 무시
    # 모든 그리드 셀에 대해 순회
    for grid in out[0]:
        # 바운딩 박스의 x, y 좌표, 너비, 높이, 그리고 객체 존재 확률을 추출
        center_x, center_y, w, h, objectness = grid[:5]

        # 객체 존재 확률이 0.5보다 크면, 해당 그리드 셀에 대한 예측을 저장
        if objectness > 0.1:
            # 이미지의 크기에 따라 바운딩 박스의 좌표와 크기를 스케일링
            center_x = int(center_x * image.shape[1])
            center_y = int(center_y * image.shape[0])
            w = int(w * image.shape[1])
            h = int(h * image.shape[0])

            # 바운딩 박스의 좌표를 계산 (YOLO는 바운딩 박스의 중심 좌표를 예측하므로, 이를 바운딩 박스의 좌측 상단 좌표로 변환해야 함)
            x = int(center_x - w / 2)
            y = int(center_y - h / 2)

            boxes.append([x, y, w, h])
            confidences.append(float(objectness))

            # 객체의 클래스 ID를 찾기 위해, 나머지 값을 확률로 간주하고 이 중에서 가장 높은 확률을 가진 클래스를 선택
            # 나머지 값을 확률로 간주하고 이 중에서 가장 높은 확률을 가진 클래스를 선택
            class_probabilities = grid[5:]
            class_id = np.argmax(class_probabilities)
            class_ids.append(class_id)
            # classes 리스트의 길이를 초과하지 않는지 확인
            # if class_id < len(classes):
            #     class_ids.append(class_id)
            # else:
            #     print(f"Warning: Predicted class id {class_id} is out of range of classes list with length {len(classes)}. Ignoring this box.")


# for out in outs:
#     for detection in out:
#         scores = detection[5:]
#         class_id = np.argmax(scores)
#         confidence = scores[class_id]
#         if confidence > 0.5:
#             # Object detected
#             center_x = int(detection[0] * width)
#             center_y = int(detection[1] * height)
#             w = int(detection[2] * width)
#             h = int(detection[3] * height)
#             # 좌표
#             x = int(center_x - w / 2)
#             y = int(center_y - h / 2)
#             boxes.append([x, y, w, h])
#             confidences.append(float(confidence))
#             class_ids.append(class_id)

# Non-Maxima Suppression 사용하여 겹치는 박스 제거
if boxes and confidences:  # 리스트가 비어 있지 않은지 확인
    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
    print(indexes)
else:
    indexes = []

# 감지된 객체에 바운딩 박스와 레이블 그리기
font = cv2.FONT_HERSHEY_PLAIN
for i in range(len(boxes)):
    if i in indexes:
        x, y, w, h = boxes[i]
        label = str(class_ids[i])
        color = colors[0]
        cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)
        cv2.putText(image, label, (x, y + 30), font, 3, color, 3)
cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# ONNX 모델 로드
net = cv2.dnn.readNetFromONNX("C:/Users/User/Desktop/InsectData/runs/detect/train31/weights/best.onnx")

classes = []
with open("C:/Users/User/Desktop/InsectData/ConvertData/HighDefinition_data/labels/classes.txt", "r") as f:
    classes = [line.strip() for line in f.readlines()]

colors = np.random.uniform(0, 255, size=(len(classes), 3))

# 이미지 로드
image = cv2.imread("C:/Users/User/Desktop/InsectData/Data/Training/rawData/TS1/PtecticusTenebrifer/Caterpillar/Cam3/Capture_20211109_15_49_04.jpg")

# 이미지 크기 조정
image = cv2.resize(image, (640, 640))

blob = cv2.dnn.blobFromImage(image, 1.0/255.0, (416, 416), (0, 0, 0), swapRB=True, crop=False)

net.setInput(blob)
outs = net.forward()

class_ids = []
confidences = []
boxes = []

for out in outs:
    for detection in out:
        scores = detection[5:]
        class_id = np.argmax(scores)
        confidence = scores[class_id]

        # 신뢰도 임계값 설정 (예: 0.5)
        if confidence > 0.5:
            center_x = int(detection[0] * image.shape[1])
            center_y = int(detection[1] * image.shape[0])
            w = int(detection[2] * image.shape[1])
            h = int(detection[3] * image.shape[0])

            x = int(center_x - w / 2)
            y = int(center_y - h / 2)

            boxes.append([x, y, w, h])
            confidences.append(float(confidence))
            class_ids.append(class_id)

indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
font = cv2.FONT_HERSHEY_PLAIN

for i in range(len(boxes)):
    if i in indexes:
        x, y, w, h = boxes[i]
        label = str(classes[class_ids[i]])
        color = colors[i]
        cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)
        cv2.putText(image, label, (x, y + 30), font, 3, color, 3)

cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

url = f'http://192.168.0.46:8888/stream'
cap = cv2.VideoCapture(url)

if not cap.isOpened():
    print("Failed to Open the IP camera Stream")
    exit()

while cap.isOpened():
    success, frame = cap.read()

    if success:
        results = model(frame)
        annotated_frame = results[0].plot()

        cv2.imshow("YOLOv8 Inference", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
